```python

S0 = 1006
I0 = 1
R0 = 0
beta = 0.2
gamma = 0.1

time_days = 100 + 2.04081633
dt = 2.04081633/4

numsteps = int(time_days / dt)
return_every = 4
```

```python
def dict_to_string(dict,  deci = 2, spaces = 7):
    '''
    Accepts a dictionary, an integer of decimals and an integer of spaces between
    the names and the values. Returns a string with the names and values of the dictionary.
    '''
    val_names = dict.keys()
    max_len = max([len(i) for i in val_names])
    spaces = spaces
    extra_space = ' '
    string = ''
    num_deci = f'{{:.{deci}f}}'
    
    for y in val_names:
        value = dict[y]
        if len(y) < max_len:
            if type(value) == tuple:
                string += y + ':' + extra_space*(max_len - len(y)) + extra_space*spaces + num_deci.format(value[0]) + '  ' + num_deci.format(value[1]) + '\n'
            else:
                string += y + ':' + extra_space*(max_len - len(y)) + extra_space*spaces + num_deci.format(value) + '\n' 
        else:
            if type(value) == tuple:
                string += y + ':' + extra_space*spaces + num_deci.format(value[0]) + '  ' + num_deci.format(value[1]) + '\n'
            else:
                string += y + ':' + extra_space*spaces + num_deci.format(value) + '\n'
    return string

def print_on_subplots(horizontal_coord, vertical_coord, values, ax, fontsize = 12, fontname = 'monospace', color = 'black'):
    '''
    Accepts coords, a string and an axis from subplots function.
    Initializes an object that plots the string on the axis at the given coords.
    Default fontsize is 12 and fontname is monospace.
    Returns nothing.

    '''
    ax.text(horizontal_coord, vertical_coord, values, color = color, fontsize = fontsize, fontname=fontname, transform=ax.transAxes)
    return
```

```python
import optuna

def objective(trial: optuna.Trial) -> float:
	bet = trial.suggest_float("betta", 1e-4,1)
	gam = trial.suggest_float("gamma", 1e-4,1)
	sim = SIR_python.integrate_system(S0, I0, R0, bet, gam, dt,numsteps, return_every)
	N = S0 + I0 + R0
	return np.sum(
        ((sim[:,1] - S_data)/N)**2 +
        ((sim[:,2] - I_data)/N)**2 +
        ((sim[:,3] - R_data)/N)**2
    )

study = optuna.create_study(direction="minimize");
study.optimize(objective, n_trials=500);

best_betta = study.best_params["betta"]
best_gamma = study.best_params["gamma"]
print(f"Best betta: {best_betta}, Best gamma: {best_gamma}")
## plot data
best_sim = SIR_python.integrate_system(S0, I0, R0, best_betta, best_gamma, dt,numsteps, return_every)

#plotting params
import seaborn as sns
sns.set_context("paper", font_scale=1.4)
sns.set_style("ticks")

plt.rcParams['axes.grid'] = True
#plt.rcParams['axes.grid.axis'] = 'y'
#plt.rcParams['axes.grid.which'] = 'major'
plt.rcParams['grid.linestyle'] = '--'
plt.rcParams['grid.linewidth'] = 1.5
plt.rcParams['figure.figsize'] = (16, 9)
plt.rcParams['figure.dpi'] = 150
plt.style.use('seaborn-v0_8')

rc = {"font.family" : "serif", 
      "mathtext.fontset" : "stix"}
plt.rcParams.update(rc)
plt.rcParams["font.serif"] = ["Times New Roman"] + plt.rcParams["font.serif"]

fig, ax = plt.subplots()
plt.plot(best_sim[:,0], best_sim[:,1], label="Best Susceptible")
plt.plot(best_sim[:,0], best_sim[:,2], label="Best Infected")
plt.plot(best_sim[:,0], best_sim[:,3], label="Best Recovered")
plt.xlabel("Time")
plt.ylabel("Number of individuals")
plt.plot(time_data, S_data, "o", label="Susceptible data")
plt.plot(time_data, I_data, "o", label="Infected data")
plt.plot(time_data, R_data, "o", label="Recovered data")

plt.grid()
plt.xlabel("# days")
plt.ylabel("# people")
plt.title("Unknown fitted SIR data")

dictio = {"Best beta":best_betta, "Best gamma":best_gamma} 
a = dict_to_string(dictio, deci = 4)
print_on_subplots(0.02,0.5,a,ax, fontsize = 10, color = "green")

plt.legend()
plt.savefig("Unknown_SIR_data_fit")

plt.show()
```
